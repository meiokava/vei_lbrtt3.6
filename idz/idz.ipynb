<font size ="4">Построение графика функции:

$$f(x, y) = sin(x) * cos(y) * (x + y)$$

в диапазоне -5 ≤ x ≤ 5, -5 ≤ y ≤ 5 и -10 ≤ z ≤ 10.
    

Результирующий график представляет собой трехмерный поверхностный график функции f(x, y) в указанном диапазоне значений x и y.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_min, x_max = -5, 5
y_min, y_max = -5, 5
z_min, z_max = -10, 10

x, y = np.meshgrid(np.linspace(x_min, x_max, 100),np.linspace(y_min, y_max, 100))
z = np.sin(x) * np.cos(y) * (x + y) 
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show() 
```
